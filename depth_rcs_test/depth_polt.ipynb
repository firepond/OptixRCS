```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# rcs_diagram("hawker_800")
rcs_file = "./a380_depth.csv"

data = np.genfromtxt(rcs_file, delimiter=",")

fig, ax = plt.subplots()
for i in range(0, 32, 4):
    depth = i
    cur_data = data[i*181:(i+1)*181, :]
    rcs = cur_data[:, 3]
    # reflectance=data[:,0]
    phi = cur_data[:, 1]
    ax.plot(phi, rcs, label="" + str(depth))

ax.legend()
ax.set_xlabel("Azimuth angle (degree)")
ax.set_ylabel("RCS (dBsm)")
fig.savefig('a380_depth.png', transparent=True)
```

```python
# rcs_diagram("hawker_800")
rcs_file = "./a380_depth.csv"

data = np.genfromtxt(rcs_file, delimiter=",")

std_data = data[31*181:(31+1)*181, 3]

result = []
for i in range(0, 32):
    cur_data = data[i*181:(i+1)*181, :]
    rcs = cur_data[:, 3]
    error = ((rcs-std_data)**2).mean()
    result.append(error)

depths = np.arange(0, 32, 1)
print(result)
result = np.array(result)

fig, ax = plt.subplots()
ax.plot(depths, result)

ax.set_xlabel("Depth")
ax.set_ylabel("Mean squared error")
# ax.show()

fig.savefig('a380_mse.png', transparent=True)
```
