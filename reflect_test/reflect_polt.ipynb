```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# rcs_diagram("hawker_800")
rcs_file = "./a380_reflect.csv"

data = np.genfromtxt(rcs_file, delimiter=",")


fig, ax = plt.subplots()
for i in range(2, 11, 2):
    refl = i / 10
    cur_data = data[i*181:(i+1)*181, :]
    rcs = cur_data[:, 3]
    # reflectance=data[:,0]
    phi = cur_data[:, 1]
    ax.plot(phi, rcs, label="" + str(refl))

ax.legend()
ax.set_xlabel("Azimuth angle (degree)")
ax.set_ylabel("RCS (dBsm)")
# ax.show()
fig.savefig('a380_refl.png', transparent=True)
```
