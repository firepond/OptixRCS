```python
import subprocess
import numpy as np
import matplotlib.pyplot as plt
from runner import Runner
from matplotlib.pyplot import imread
```

```python
phi_runner = Runner()
model_name = "corner_reflector"
model_file = "C://development//optix//OptixRCS//resources//" + model_name+".obj"
csv_file_vs = "C://development//optix//OptixRCS//output//"+ model_name + '_rcs_phi.csv'
csv_file = "../output/" +model_name + '_rcs_phi.csv'

phi_runner.model_file = model_file
phi_runner.csv_file = csv_file_vs
phi_runner.freq=3E9
phi_runner.phi_start=0
phi_runner.phi_end=90
phi_runner.phi_interval=1
phi_runner.theta_start=60
phi_runner.theta_end=60
phi_runner.rays_per_wavelength=100
phi_runner.polarization = "HH"
phi_runner.generate_json()
phi_runner.run()
```

should be:

rcs max:  26.0335

rcs min:  14.4096

```python
csv = np.genfromtxt (csv_file, delimiter=",")
csv=csv[~np.isnan(csv[:,3])]
# Extract phi and rcs columns from the DataFrame
freq_values = csv[:,0]
phi_values = csv[:,1]
theta_values = csv[:,2]
rcs_values = csv[:,3]
print("rcs max: ", max(rcs_values))
print("rcs min: ", min(rcs_values))
```

```python
# Create the plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.plot(phi_values, rcs_values)
plt.xlabel('Incident angle (degree)')  # X-axis label
plt.ylabel('RCS (dBsm)')  # Y-axis label
# plt.title('RCS vs Phi')  # Plot title
plt.show()  # Display the plot
```
